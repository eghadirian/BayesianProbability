```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
import arviz as az
from pymc3.ode import DifferentialEquation
import theano.tensor as tt
import theano
from scipy.integrate import odeint
from scipy.integrate import trapz
from scipy.stats.mstats import mquantiles
```

```python
class COVID_data():
    def __init__(self, country='US', Population = 328.2e6,\
                 data_begin = '1/22/20', data_end = '11/1/20',\
                 cases='time_series_covid19_confirmed_global.csv'):
        confirmed_cases = pd.read_csv(cases, sep=',')
        self.N = Population
        self.to_date_cases_us = confirmed_cases.loc[confirmed_cases\
                  ["Country/Region"] == country].transpose().iloc[4:,:]
        self.to_date_cases_us.columns = ['cases']
        self.to_date_cases_us = self.to_date_cases_us
        self.cases_us = self.to_date_cases_us[data_begin:data_end]
        self.cases_us_inc = pd.DataFrame({'cases':np.gradient(covid_data.cases_us['cases'])})
```

```python
covid_data = COVID_data()
```

```python
class SIR_model():
    def __init__(self, covid_data) :
        self.covid_data = covid_data
        self.setup_SIR_model()
 
    def setup_SIR_model(self):
        self.time_range = np.arange(0,len(self.covid_data.cases_us),1)
        self.I0 = self.covid_data.cases_us['cases'].to_numpy()[0]
        self.S0 = (self.covid_data.N - self.I0)
```

```python
sir = SIR_model(covid_data)
```

```python
def SIR_non_normalized(y, t, p):
    ds = -p[0] * y[0] * y[1]/sir.covid_data.N
    di = p[0] * y[0] * y[1]/sir.covid_data.N -  p[1] * y[1]
    return [ds, di]

ode_model = DifferentialEquation(func=SIR_non_normalized, \
                                 times=sir.time_range,\
                                 n_states=2, n_theta=2, t0=0)
```

```python
cum_cases = np.zeros_like(sir.time_range)
with pm.Model() as model:
    sigma = pm.HalfCauchy('sigma', 2.)
    lam = pm.Lognormal('lambda', np.log(1), 0.5)
    mu = pm.Lognormal('mu', np.log(1), 0.2)
    ode_solution = ode_model(y0=[sir.S0, sir.I0], theta=[lam, mu])
    Y = pm.Normal('Y', mu=cum_cases, \
                    sigma=sigma, observed=sir.covid_data.cases_us_inc['cases'])
    R0 = pm.Deterministic('R0', lam/mu)
    trace = pm.sample(3000, tune=1000, random_seed=42, step=pm.Metropolis())
```

```python
lambda_samples = trace['lambda']
mu_samples = trace['mu']
R0_samples = trace['R0']
```

```python
plt.figure()
plt.hist(lambda_samples, histtype='stepfilled', bins=35, alpha=0.85,
        label=r"posterior of $\lambda$")
plt.legend()
plt.figure()
plt.hist(mu_samples, histtype='stepfilled', bins=35, alpha=0.85,
        label=r"posterior of $\mu$")
plt.legend()
plt.figure()
plt.hist(R0_samples, histtype='stepfilled', bins=35, alpha=0.85,
        label=r"posterior of R0")
plt.legend()
```

```python
az.plot_trace(trace, var_names='mu')
az.plot_trace(trace, var_names='lambda')
az.plot_trace(trace, var_names='R0')
```

```python
print('The basic reproduction number is estimated as {:3.2}, stddv of {:4.3}'.format(R0_samples.mean(), R0_samples.std()))
```
